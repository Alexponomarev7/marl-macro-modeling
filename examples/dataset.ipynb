# Analyze dataset

```python
from pathlib import Path

dataset_path = Path("/Users/aponomarev/marl_experiments/c205cbb48093907e72c373c15cdd77df/dataset/train")
```

```python
import json
with open(dataset_path / "metadata.json") as f:
    metadata = json.load(f)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

_, ax = plt.subplots(len(metadata), 2, figsize=(10, 20))
for i, m in enumerate(metadata):
    p = pd.read_parquet(m["output_dir"])

    ax[i, 0].plot(p['state'].map(lambda x: x['Capital']).values)
    ax[i, 0].set_title(m['env_name'])
    ax[i, 0].set_xlabel("Step")
    ax[i, 0].set_ylabel("Capital")

    ax[i, 1].plot(p['info'].map(lambda x: x['consumption']).values)
    ax[i, 1].set_title(m['env_name'])
    ax[i, 1].set_xlabel("Step")
    ax[i, 1].set_ylabel("Consumption")
```
