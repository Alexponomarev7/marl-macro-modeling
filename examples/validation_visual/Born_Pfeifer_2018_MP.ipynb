```python
import pandas as pd
pd.set_option('display.max_columns', 500)

raw_data = pd.read_csv("/Users/annaevseeva/Downloads/MARL Dynare/Born_Pfeifer_2018_MP.csv")
print(raw_data.columns.to_list())
raw_data.head()
```

```python
column_mapping = {
    'price_inflation': 'Price Inflation',
    'output_gap': 'Output Gap',
    'natural_output': 'Natural Output',
    'output': 'Output',
    'output_deviation': 'Output Deviation From Steady State',
    'natural_interest_rate': 'Natural Interest Rate',
    'real_interest_rate': 'Real Interest Rate',
    'nominal_interest_rate': 'Nominal Interest Rate',
    'hours_worked': 'Hours Worked',
    'real_money_stock': 'Real Money Stock',
    'money_growth_annual': 'Money Growth Annualized',
    'nominal_money_stock': 'Nominal Money Stock',
    'monetary_shock': 'AR(1) Monetary Policy Shock Process',
    'technology_shock': 'AR(1) Technology Shock Process',
    'annual_real_rate': 'Annualized Real Interest Rate',
    'annual_nominal_rate': 'Annualized Nominal Interest Rate',
    'annual_natural_rate': 'Annualized Natural Interest Rate',
    'annual_inflation': 'Annualized Inflation Rate',
    'preference_shock': 'AR(1) Preference Shock Process',
    'price_level': 'Price Level',
    'nominal_wage': 'Nominal Wage',
    'consumption': 'Consumption',
    'real_wage': 'Real Wage',
    'real_wage_gap': 'Real Wage Gap',
    'wage_inflation': 'Wage Inflation',
    'natural_wage': 'Natural Real Wage',
    'price_markup': 'Markup',
    'annual_wage_inflation': 'Annualized Wage Inflation Rate'
}


raw_data = raw_data.rename(columns=column_mapping)


print(raw_data.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

columns = ["Price Inflation", "Output Gap", "Natural Output", "Output", "Output Deviation From Steady State", "Natural Interest Rate", "Consumption"]
# columns = ["y", "c", "k", "invest"]


fig, axes = plt.subplots(nrows=len(columns), ncols=1, figsize=(13.33, 7.5 * len(columns)))

for idx, column in enumerate(columns):
    
    column_values = raw_data[column].to_numpy()
    
    steps = np.arange(len(column_values))
    
    ax = axes[idx]
    
    ax.plot(steps, column_values, marker='o', color='blue', label=f"Dynamics of {column}")
    
    
    ax.legend(fontsize=12)
    ax.set_xlabel("Step", fontsize=14)
    ax.set_ylabel(f"{column}", fontsize=14)
    ax.set_xticks(steps[::4])
    # ax.set_yticks(np.arange(np.min(column_values), np.max(column_values)))
    ax.set_xticklabels(ax.get_xticks(), rotation=90)
    ax.grid(True)
```
