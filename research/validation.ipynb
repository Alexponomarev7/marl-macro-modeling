```python


action = env.action_space.sample()
sim_state, sim_reward, _, _, sim_info = env.step(action)
env.reset()
env.step(action)
ana_state, ana_reward, _, _, ana_info = env.analytical_step()
result = {
    "Simulated_Reward": sim_reward,
    "Analytical_Reward": ana_reward,
    "Reward_Diff": abs(sim_reward - ana_reward),
    "Simulated_Output": sim_info.get("output"),
    "Analytical_Output": ana_info.get("output"),
    "Output_Diff": abs(sim_info.get("output") - ana_info.get("output")),

    # new
    "Sim_Consumption": sim_info.get("consumption"),
    "Ana_Consumption": ana_info.get("consumption"),
    "Consumption_Diff": abs(sim_info.get("consumption", 0) - ana_info.get("consumption", 0)),

    "Sim_Capital": sim_info.get("capital"),
    "Ana_Capital": ana_info.get("capital"),
    "Capital_Diff": abs(sim_info.get("capital", 0) - ana_info.get("capital", 0)),
}

result
```
